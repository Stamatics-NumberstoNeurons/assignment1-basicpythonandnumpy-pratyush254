### Exercise 41 — Create New Column

**Question:** Create a new column for **volume** in `iris_2d`, where:

> volume = (π × petal_length × sepal_length²) / 3

*# Make sure you have run Exercise 27 first.*


```python
# Write your code here:
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/iris.csv"

iris_data = pd.read_csv(url, header=None)

iris_2d = iris_data.iloc[:, :4].to_numpy()

sepal_length = iris_2d[:, 0]
petal_length = iris_2d[:, 2]

volume = (np.pi * petal_length * (sepal_length ** 2)) / 3

iris_2d = np.column_stack((iris_2d, volume))

print(iris_2d)
```

**Question:** Convert `arr_2d` (an array of arrays) into a flat 1D array.


```python
# Input — do not modify
arr1 = np.arange(3)
arr2 = np.arange(3, 7)
arr3 = np.arange(7, 10)
arr_2d = np.array([arr1, arr2, arr3], dtype=object)
```

```python
# Write your code here:
import numpy as np

arr1 = np.arange(3)
arr2 = np.arange(3, 7)
arr3 = np.arange(7, 10)

arr_2d = np.array([arr1, arr2, arr3], dtype=object)

flat_array = np.concatenate(arr_2d)

print(flat_array)
```
